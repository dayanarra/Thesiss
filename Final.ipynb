# Cleaning tweets

The cleaning of the data will involve removing some punctuation, hyperlinks and hashtags.

```python
# removing only some punctuation, because the VADER package takes into account ! and ? which influence the polarity of the text

import re
punct = '"#$%&\'()*+,-./:;<=>@[\\]^_`{|}~'
punct
```

Below is the function to clean tweets. Initial cleaning will involve removing some punctuation, removing https links, removing /r & /n, and removing hashtags in front of words.

- The arguments in comments can further clean the text by removing numbers, emojis and putting everything in lowercase. But the VADER algorithm recognizes emojis, and caps lock, and ! ? as expressing emotions. Thus, initially they will be kept, in the modeling stage further cleaning can be applied if the performance is unsatisfactory.

```python
# Function to clean tweets
def cleantwt(text):
    text  = "".join([char for char in text if char not in punct]) #removes punctuation
    #text = re.sub('[0-9]+', '', text) #removes numbers
    text = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", text) #Remove http links
    #text = " ".join(tweet.split())
    text = text.replace('\r', '').replace('\n', ' ').replace('\n', ' ') #.lower() #remove \n and \r and lowercase
    text = text.replace("#", "").replace("_", " ")
    #text = re.sub(r'[^\x00-\x7f]',r'', text) #remove non utf8/ascii characters such as '\x9a\x91\x97\x9a\x97' this removes emojis
    return text

bittweets['clean_tweet'] = bittweets['content'].apply(lambda x: cleantwt(x))
```

# VADER

Applying the sentiment analyzer on the tweets, by first computing the 4 VADER scores(compound, positive, negative, and neutral).


```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import numpy as np
analyzer = SentimentIntensityAnalyzer()
bittweets['compound'] = [analyzer.polarity_scores(x)['compound'] for x in bittweets['clean_tweet']]
bittweets['neg'] = [analyzer.polarity_scores(x)['neg'] for x in bittweets['clean_tweet']]
bittweets['neu'] = [analyzer.polarity_scores(x)['neu'] for x in bittweets['clean_tweet']]
bittweets['pos'] = [analyzer.polarity_scores(x)['pos'] for x in bittweets['clean_tweet']]
```

```python
bittweets
```

Adding a column 'sentiment' where each tweet is classified as either positive, negative or neutral.

```python

result = []
for value in bittweets["compound"]:
    if value >= 0.05:
        result.append("Positive")
    elif value <= (-0.05):
        result.append("Negative")
    else:
        result.append("Neutral")

bittweets["Sentiment"] = result
```
